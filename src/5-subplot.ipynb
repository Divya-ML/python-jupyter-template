```python
# TASK: Create a 2x2 subplot layout.

import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2)

# Plot some data on each subplot
axs[0, 0].plot([1, 2, 3], [1, 4, 9])  # Top-left subplot
axs[0, 0].set_title('Plot 1')

axs[0, 1].bar([1, 2, 3], [3, 6, 9])  # Top-right subplot
axs[0, 1].set_title('Plot 2')

axs[1, 0].scatter([1, 2, 3], [1, 2, 3])  # Bottom-left subplot
axs[1, 0].set_title('Plot 3')

axs[1, 1].pie([30, 30, 20, 20], labels=['A', 'B', 'C', 'D'])  # Bottom-right subplot
axs[1, 1].set_title('Plot 4')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2)

# 1st subplot: Line chart
axs[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axs[0, 0].set_title('Line Chart')

# 2nd subplot: Bar chart
axs[0, 1].bar([1, 2, 3, 4], [2, 4, 6, 8])
axs[0, 1].set_title('Bar Chart')

# 3rd subplot: Scatter plot
axs[1, 0].scatter([1, 2, 3, 4], [1, 2, 3, 4])
axs[1, 0].set_title('Scatter Plot')

# 4th subplot: Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [25, 30, 20, 25]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
