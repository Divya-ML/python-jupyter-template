```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
#TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].

import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create a histogram
plt.hist(data, bins=5, edgecolor='black', alpha=0.7)

# Add title and labels
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create a histogram with custom color and edgecolor
plt.hist(data, bins=5, edgecolor='black', color='skyblue', alpha=0.7)

# Add title and labels
plt.title('Customized Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
