### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Generate y values as the square of x
y = x**2

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Line Plot of x vs x^2')  # Add a title
plt.xlabel('x')  # Label for the x-axis
plt.ylabel('x^2')  # Label for the y-axis
plt.grid(True)  # Add a grid

# Show the plot
plt.show()
```
