```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create the bar plot
plt.bar(categories, values)

# Add title and labels
plt.title('Bar Plot of Categories vs Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Use different colors for each bar and add a title to the plot.

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

# Create the bar plot with custom colors
plt.bar(categories, values, color=colors)

# Add title and labels
plt.title('Bar Plot of Categories vs Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```
